```python

def decode_waybill(record, layout):
    decoded_record = {}
    for field in layout:
        start_idx = field["start"] - 1
        end_idx = field["end"]
        value = record[start_idx:end_idx].strip()
        decoded_record[field["name"]] = record[start_idx:end_idx].strip()

        if field["name"] == "Car Ownership Code":
            value = decode_car_ownership(value)
        elif field["name"] == "Trailer/Container Ownership Code":
            value = decode_trailer_ownership(value)
        elif field["name"] == "Hazardous/Bulk Material":
            value = decode_hazardous_bulk(value)
        elif field["name"] == "Origin BEA Area" or field["name"] == "Termination BEA Area":
            value = bea_codes.get(value, "Unknown BEA Code")
        decoded_record[field["name"]] = value

    return decoded_record


def decode_car_ownership(code):
    return {
        'P': 'Privately-owned car',
        'T': 'Trailer Train owned car'
    }.get(code, 'Railroad-owned car')

def decode_trailer_ownership(code):
    return {
        'P': 'Privately-owned Trailer/Container',
        'R': 'Railroad-owned Trailer/Container'
    }.get(code, 'Unknown')

def decode_hazardous_bulk(code):
    return {
        'B': 'Bulk, non-hazardous material in Boxcar',
        'H': 'Hazardous material in car'
    }.get(code, 'Neither')

    return decoded_record

```

```python
import pandas as pd

data_file = 'PublicUseWaybillSample2022.txt'

def process_large_file(file_path, record_length=248, chunk_size=10000):
    decoded_records = []
    current_record_lines = []
    with open(file_path, 'r') as file:
        while True:
            lines = file.readlines(chunk_size)
            if not lines:
                break
            for line in lines:
                decoded_record = decode_waybill(line, waybill_layout)
                decoded_records.append(decoded_record)
            current_record_lines = []
    return pd.DataFrame(decoded_records)

df = process_large_file(data_file, record_length=247, chunk_size=10000)
df
```
